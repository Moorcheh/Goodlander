```python
N = 100
```

```python
import random

brownianMotion = [0]

for t in range(N):
    currentLocation = brownianMotion[-1]
    step = random.normalvariate(0, 1)
    nextLocation = currentLocation + step
    brownianMotion.append(nextLocation)
```

```python
import matplotlib.pyplot as plt

plt.plot(brownianMotion)
```

```python
import math

drift = 0
vol = 100 / 100
initialPrice = 100

def price(t, brownianMotion):
    if type(t) is not int: return
    volConverted = vol / N
    driftGrowth = math.exp(drift * t)
    volDecay = math.exp(-volConverted**2/2 * t)
    brownianShock = math.exp(volConverted * brownianMotion)
    return initialPrice * driftGrowth * volDecay * brownianShock

prices = [price(t, brownianMotion) for t, brownianMotion in enumerate(brownianMotion)]
```

```python
plt.plot(prices)
```
